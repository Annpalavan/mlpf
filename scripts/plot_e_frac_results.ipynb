```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
plt.ioff()
in_dir = "/eos/user/g/gkrzmanc/temporary_results_inference1.pkl"
in_dir2 = "/eos/user/g/gkrzmanc/temporary_results_inference1_part2.pkl"

results = pickle.load(open(in_dir, "rb"))
results2 = pickle.load(open(in_dir2, "rb"))
results.update(results2)
radiuses = sorted(list(results.keys()))
print(radiuses)
radiuses = radiuses[:8]
#print(radiuses)
#radiuses = radiuses[:4]

results_hgcal = pickle.load(open("/eos/user/g/gkrzmanc/summ_results/frac_energy_plots/hgcal_model/result_e_frac11.pkl", "rb"))
```

```python
loss_e_fracs = [np.median(np.array(results[r]["loss_e_fracs"])) for r in radiuses]
loss_e_fracs_true = [np.median(np.array(results[r]["loss_e_fracs_true"])) for r in radiuses]

loss_e_fracs_hgcal = [np.median(np.array(results_hgcal[r]["loss_e_fracs"])) for r in radiuses]
loss_e_fracs_true_hgcal = [np.median(np.array(results_hgcal[r]["loss_e_fracs_true"])) for r in radiuses]
```

```python
fig, ax = plt.subplots()
ax.plot(radiuses, loss_e_fracs, ".--", label="loss_e_fracs")
ax.plot(radiuses, loss_e_fracs_true, ".--", label="loss_e_fracs_true")
ax.plot(radiuses, loss_e_fracs_hgcal, ".--", label="loss_e_fracs_hgcal_impl")
ax.plot(radiuses, loss_e_fracs_true_hgcal, ".--", label="loss_e_fracs_true_hgcal_impl")

ax.set_xlabel("radius")
ax.set_ylabel("loss")
ax.legend()
ax.grid()
ax.set_ylim([-1, 1.15])

fig.savefig("/eos/user/g/gkrzmanc/temp_fig.png")
# display fig
from IPython.display import Image
Image(filename='/eos/user/g/gkrzmanc/temp_fig.png')
```
